#### 2. Opening csv file from url

- how this works is

1. Your code is now the client that hits the url where the csv is present and with the help of requests.get fetches the content of that file and stores in req. 

2. StringIO stores the text from the file into data \n
   
4. Now you can do read_csv on data

```python
import requests
from io import StringIO
url = "https://raw.githubusercontent.com/cs109/2014_data/master/countries.csv"
headers = {"User-Agent": "Mozilla/5.0 (Macintosh; Intel Mac OS X 10.14; rv: 66.0) Gecko/20100101 Firefox/66.0"}
req = requests.get(url, headers=headers)
data = StringIO(req. text)
pd. read_csv(data)
```

#### 14.Converters
- using this you can convert the names of values of a specific columns using a function
- ex in IPL dataset team names are written completely..instead if I want to use shortforms for the teams, what I'll do is:
  1. Prepare a function 'rename' and write the code for renaming the teams.
  2. Then I'll pass the function in converters parameter.
  3. Format of converters: converters = {'*col_name'*,*func_name* }

```python
def rename(name):
    if name == 'Royal Challengers Bangalore':
        return 'RCB'
    else:
        return name
```

```python
rename('Royal Challengers Bangalore')
```

```python
#befor converters
pd.read_csv("Datasets/IPL Matches 2008-2020.csv")
```

```python
pd.read_csv("Datasets/IPL Matches 2008-2020.csv",converters = {'team1':rename, 'team2':rename})
```

#### 15. na_values parameter
- It is used to identify what type values should be considered NaN
- Sometimes instead of NaN, datasets contain '-', 'na', such values
- But these values are not identified as NaN by pandas.
- So by using na_values parameter you can tell pandas which values to consider NaN

```python
#just for the sake of understanding i am taking male as NaN
pd.read_csv("Datasets/aug_train.csv", na_values = ['Male'])
```
